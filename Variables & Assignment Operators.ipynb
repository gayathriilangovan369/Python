```python
print("Calculating Area of Triangle")
Base=int(input("Please enter base"))
Height=int(input("please enter height"))
Area=(Base*Height)/2
print("Base is", Base)
print("Height is", Height)
print("Area is ",Area)
```

```python
print("Calculating Area of Circle")
Radius=float(input("Please enter radius"))
Area=3.14*Radius**2
print("Radius is", Radius)
print("Area is ",Area)
```

```python
print("Calculating Area of Parallelogram")
Length=int(input("Please enter Length"))
Width=int(input("please enter Width"))
Area=Length*Width
print("Length is", Length)
print("Width is",Width )
print("Area is ",Area)
```

```python
print("Converting Celsius to Fahrenheit")
Celsius=int(input("Please enter Celsius"))
Fahrenheit=(Celsius * (9/5)) + 32
print("Celsius is", Celsius)
print("Fahrenheit is",Fahrenheit,"°F" )
```
